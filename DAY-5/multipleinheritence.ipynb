```python
class Product:
    def review(self):
        print("product cutomer review")


class Phone(Product):
    def __init__(self,price,brand,camera):
        print("inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
    def buy(self):
        print("buying a phone")

class SmartPhone(Phone):
    pass






s=SmartPhone(200000,"samsung",3)
s.buy()
s.review()
    
```

```python
#multiple inheritence

#multiple parent
#parent
class Phone:
    def __init__(self,price,brand,camera):
        print("inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("buying a phone")
 #parent       
class Product:
    def __init__(self):
        pass
    def review(self):
        print("product customerreview")
        
#child
class SmartPhone(Phone,Product):
    pass



s=SmartPhone(10000,"apple","3")
s.buy()
s.review()



        
        


```
