```python
class Person:
    def __init__(self,name,country):#parameters
        self.name=name
        self.country=country 
        
```

```python
p1=Person("satheesh","india")
p2=Person("ammulu","uk")
p3=Person("mouni","germany")
```

```python
l=[p1,p2,p3]
print(l)
```

```python
for i in l:
    print(i.name,i.country)
```
